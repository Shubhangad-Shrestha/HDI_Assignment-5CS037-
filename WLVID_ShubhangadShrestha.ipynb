## Problem 1A – Single Year HDI Exploration (2022)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#Task 1: Extract Latest Year (2022)
df = pd.read_csv("Human_Development_Index_Dataset.csv", encoding='latin1')
df.head()
```

```python

# List all unique years
print(df['year'].unique())
# Filter rows where year is 2022
hdi_2022_df = df[df['year'] == 2022]

#Task 2: Data Exploration

# Show first 10 rows to check
hdi_2022_df.head(10)
```

```python
# Check shape (rows, columns)
print("Rows, Columns:", hdi_2022_df.shape)

# Check column names and data types
print(hdi_2022_df.dtypes)
```

```python
#Task 3: Missing Values & Data Cleaning

# Check missing values per column
print("Missing values:\n", hdi_2022_df.isnull().sum())

# Check duplicate rows
print("Duplicate rows:", hdi_2022_df.duplicated().sum())

# Remove duplicates if any
hdi_2022_df = hdi_2022_df.drop_duplicates()
```

```python
# Check all column names exactly as pandas sees them
print(df.columns)
```

```python
# Convert HDI and GNI per capita to numeric (just in case there are weird strings)
hdi_2022_df['hdi'] = pd.to_numeric(hdi_2022_df['hdi'], errors='coerce')
hdi_2022_df['gross_inc_percap'] = pd.to_numeric(hdi_2022_df['gross_inc_percap'], errors='coerce')
# Missing values are retained (not dropped) to avoid unintended data loss,
# as HDI rankings depend on complete country information.
```

```python
# Task 4: Basic Statistics (HDI – 2022)
print("Mean HDI:", hdi_2022_df['hdi'].mean())
print("Median HDI:", hdi_2022_df['hdi'].median())
print("Standard Deviation HDI:", hdi_2022_df['hdi'].std())

# Country with highest HDI
print("Highest HDI country:", hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), 'country'])

# Country with lowest HDI
print("Lowest HDI country:", hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), 'country'])
```

```python
#Task 5: Filtering and Sorting
# Sort the filtered dataset by Gross National Income per Capita (descending)
high_hdi_df = hdi_2022_df[hdi_2022_df['hdi'] > 0.800].sort_values(by='gross_inc_percap', ascending=False)
high_hdi_df.head(10)
```

```python
#Task 6: Adding HDI Category Column

# Function to classify HDI values into official UNDP categories
def categorize_hdi(hdi):
    if hdi < 0.550:
        return 'Low'
    elif 0.550 <= hdi <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi <= 0.799:
        return 'High'
    else:
        return 'Very High'

# Apply HDI category classification and Verify classification
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(categorize_hdi)
hdi_2022_df[['country', 'hdi', 'HDI_Category']].head(10)

# Save cleaned CSV
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

# Problem 1B — HDI Trend Analysis (2020–2022)


```python
#Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task 1: Data Extraction and Saving

# Filter dataset for years 2020, 2021, 2022
hdi_2020_2022_df = df[df['year'].isin([2020, 2021, 2022])]

# Check the filtered data
hdi_2020_2022_df.head()
```

```python
#Task 2: Data Cleaning

# Convert HDI and GNI per capita to numeric
hdi_2020_2022_df['hdi'] = pd.to_numeric(hdi_2020_2022_df['hdi'], errors='coerce')
hdi_2020_2022_df['gross_inc_percap'] = pd.to_numeric(hdi_2020_2022_df['gross_inc_percap'], errors='coerce')

# Check missing values
print("Missing values:\n", hdi_2020_2022_df.isnull().sum())

# Drop duplicates if any
hdi_2020_2022_df = hdi_2020_2022_df.drop_duplicates()
# Justification:
# - Numeric conversion handles special characters like “–”
# - Duplicate rows removed to avoid bias
# - Missing values retained to preserve country-level trends

```

```python
#Task 3A: Line Chart — HDI Trend (Country-Level)

# Select 5 countries
countries = ['Nepal', 'India', 'Bangladesh', 'Pakistan', 'Sri Lanka']
trend_df = hdi_2020_2022_df[hdi_2020_2022_df['country'].isin(countries)]

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(data=trend_df, x='year', y='hdi', hue='country', marker='o')
plt.title("HDI Trend (2020–2022) for Selected South Asian Countries")
plt.ylabel("HDI")
plt.xlabel("Year")
plt.xticks([2020, 2021, 2022])
plt.legend(title='Country')
plt.show()
# Most selected countries show gradual HDI improvement, though growth rates differ.

```

```python
# Average HDI per country
country_avg_hdi = hdi_2020_2022_df.groupby('country')['hdi'].mean().sort_values(ascending=False).reset_index()

# Top 5 countries
top5 = country_avg_hdi.head(5)

# Plot
plt.figure(figsize=(8,6))
sns.barplot(data=top5, x='hdi', y='country', palette='viridis')
plt.title("Top 5 Countries by Average HDI (2020–2022)")
plt.xlabel("Average HDI")
plt.ylabel("Country")
plt.show()
```

```python
# Box plot for HDI spread across 2020, 2021, 2022
plt.figure(figsize=(8,6))
sns.boxplot(data=hdi_2020_2022_df, x='year', y='hdi', palette='coolwarm')
plt.title("HDI Distribution Across 2020–2022")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.show()
```

```python
# Check if the column exists
if 'gross_inc_percap' in hdi_2020_2022_df.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=hdi_2020_2022_df, x='gross_inc_percap', y='hdi', hue='year', palette='viridis')
    plt.title("HDI vs GNI per Capita (2020–2022)")
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.legend(title="Year")
    plt.show()
else:
    print("GNI per Capita column not available in the dataset.")
```

```python
# Task 4: Short Analysis
# Pivot to get HDI per country for 2020 and 2022
hdi_change = hdi_2020_2022_df[hdi_2020_2022_df['year'].isin([2020,2022])]
hdi_change = hdi_change.pivot(index='country', columns='year', values='hdi').reset_index()
hdi_change['HDI_diff'] = hdi_change[2022] - hdi_change[2020]

# Top 5 improvements
top_improvements = hdi_change.sort_values(by='HDI_diff', ascending=False).head(5)
print("Top 5 countries with greatest HDI improvement (2020–2022):")
print(top_improvements[['country','HDI_diff']])

# Any declines?
declines = hdi_change[hdi_change['HDI_diff'] < 0]
print("\nCountries with HDI decline:")
print(declines[['country','HDI_diff']])
```

# Problem 2 – Advanced HDI Exploration (South Asia)


```python
# Load the cleaned 2020-2022 dataset
hdi_df = pd.read_csv("HDI_problem1B.csv")
hdi_df.head()
```

```python
# Task 1: Create South Asia Subset


south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

hdi_southasia_df = hdi_df[hdi_df['country'].isin(south_asia_countries)]
hdi_southasia_df.to_csv("HDI_SouthAsia.csv", index=False)
hdi_southasia_df.head()
# Insight: Only South Asian countries are retained for advanced analysis

```

```python
# Task 2: Composite Development Score

# Create Composite Score = 0.3 * Life Expectancy Index + 0.3 * GNI per Capita Index
hdi_southasia_df['Composite_Score'] = 0.3*hdi_southasia_df['life_expectancy'] + \
                                      0.3*hdi_southasia_df['gross_inc_percap']

# Rank by Composite Score
hdi_southasia_df = hdi_southasia_df.sort_values(by='Composite_Score', ascending=False)

# Top 5 countries plot
plt.figure(figsize=(8,6))
sns.barplot(data=hdi_southasia_df.head(5), x='Composite_Score', y='country', palette='viridis')
plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.show()
# Insight: Compare rankings with HDI; differences highlight countries where GNI or life expectancy shifts overall development ranking.
```

```python
# Task 3: Outlier Detection

# Function to detect outliers
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return series[(series < lower) | (series > upper)]
# Detect HDI and GNI per Capita outliers

hdi_outliers = detect_outliers(hdi_southasia_df['hdi'])
gni_outliers = detect_outliers(hdi_southasia_df['gross_inc_percap'])

# Scatter plot with outliers
plt.figure(figsize=(8,6))
sns.scatterplot(data=hdi_southasia_df, x='gross_inc_percap', y='hdi')
sns.scatterplot(x=gni_outliers, y=hdi_southasia_df.loc[gni_outliers.index, 'hdi'], color='red', label='Outliers')
plt.title("HDI vs GNI per Capita (South Asia) with Outliers")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()
plt.show()
```

```python

# Task 4: Exploring Metric Relationships

# Correlations
print("Correlation with HDI:")
print(hdi_southasia_df[['gender_development', 'life_expectancy', 'hdi']].corr()['hdi'])

# Scatter plots
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.scatterplot(data=hdi_southasia_df, x='gender_development', y='hdi')
plt.title("HDI vs Gender Development Index")

plt.subplot(1,2,2)
sns.scatterplot(data=hdi_southasia_df, x='life_expectancy', y='hdi')
plt.title("HDI vs Life Expectancy")
plt.show()
```

```python
# Task 5: Gap Analysis

# Compute GNI-HDI Gap
hdi_southasia_df['GNI_HDI_Gap'] = hdi_southasia_df['gross_inc_percap'] - hdi_southasia_df['hdi']

# Top 3 positive gaps
top_positive = hdi_southasia_df.nlargest(3, 'GNI_HDI_Gap')

# Top 3 negative gaps
top_negative = hdi_southasia_df.nsmallest(3, 'GNI_HDI_Gap')

plt.figure(figsize=(8,6))
sns.barplot(x='GNI_HDI_Gap', y='country', data=top_positive, palette='Greens')
plt.title("Top 3 Positive GNI-HDI Gaps")
plt.xlabel("GNI-HDI Gap")
plt.ylabel("Country")
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(x='GNI_HDI_Gap', y='country', data=top_negative, palette='Reds')
plt.title("Top 3 Negative GNI-HDI Gaps")
plt.xlabel("GNI-HDI Gap")
plt.ylabel("Country")
plt.show()
# Insight: Positive gaps indicate high income but lower HDI; negative gaps indicate HDI higher than income suggests.
```


# Problem 3 – omparative Regional Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the multi-year HDI dataset
hdi_df = pd.read_csv("HDI_problem1B.csv")
hdi_df.head()
```

```python
#Task 1: Create Middle East Subset

# South Asia countries
south_asia = ["Afghanistan", "Bangladesh", "Bhutan", "India",
              "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Middle East countries
middle_east = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
               "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
               "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

# Filter datasets
hdi_south_asia = hdi_df[hdi_df['country'].isin(south_asia)]
hdi_middle_east = hdi_df[hdi_df['country'].isin(middle_east)]

# Save for later reference (optional)
hdi_south_asia.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
hdi_middle_east.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

```python
# Task 2: Descriptive Statistics

# South Asia stats
sa_mean = hdi_south_asia['hdi'].mean()
sa_std = hdi_south_asia['hdi'].std()
print("South Asia HDI Mean:", sa_mean)
print("South Asia HDI Std Dev:", sa_std)

# Middle East stats
me_mean = hdi_middle_east['hdi'].mean()
me_std = hdi_middle_east['hdi'].std()
print("Middle East HDI Mean:", me_mean)
print("Middle East HDI Std Dev:", me_std)
```

```python


# Task 3: Top and Bottom Performers

# Top 3 and bottom 3 in South Asia
sa_top = hdi_south_asia.groupby('country')['hdi'].mean().sort_values(ascending=False).head(3)
sa_bottom = hdi_south_asia.groupby('country')['hdi'].mean().sort_values().head(3)

# Top 3 and bottom 3 in Middle East
me_top = hdi_middle_east.groupby('country')['hdi'].mean().sort_values(ascending=False).head(3)
me_bottom = hdi_middle_east.groupby('country')['hdi'].mean().sort_values().head(3)

print("South Asia Top 3:\n", sa_top)
print("South Asia Bottom 3:\n", sa_bottom)
print("Middle East Top 3:\n", me_top)
print("Middle East Bottom 3:\n", me_bottom)

# Optional: plot comparison
plt.figure(figsize=(10,6))
sns.barplot(x=sa_top.index, y=sa_top.values, color='skyblue', label='SA Top')
sns.barplot(x=sa_bottom.index, y=sa_bottom.values, color='lightcoral', label='SA Bottom')
plt.title("South Asia Top & Bottom 3 HDI Countries")
plt.ylabel("Average HDI (2020–2022)")
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x=me_top.index, y=me_top.values, color='skyblue', label='Top')
sns.barplot(x=me_bottom.index, y=me_bottom.values, color='lightcoral', label='Bottom')
plt.title("Middle East Top & Bottom 3 HDI Countries")
plt.ylabel("Average HDI (2020–2022)")
plt.legend()
plt.show()

# Insight: Clear visual distinction between highest and lowest performing countries.

```

```python
# Task 4: Metric Comparisons

metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']
avg_sa = hdi_south_asia[metrics].mean()
avg_me = hdi_middle_east[metrics].mean()

df_compare = pd.DataFrame({'South Asia': avg_sa, 'Middle East': avg_me})
df_compare.plot(kind='bar', figsize=(10,6))
plt.title("Regional Comparison of Key HDI Metrics (2020–2022)")
plt.ylabel("Average Value")
plt.xticks(rotation=0)
plt.show()

# Insight: Identify which metric shows the greatest disparity between regions.
```

```python
# Task 5: HDI Disparity

# HDI Disparity: Range & CV
sa_range = hdi_south_asia['hdi'].max() - hdi_south_asia['hdi'].min()
sa_cv = hdi_south_asia['hdi'].std() / hdi_south_asia['hdi'].mean()

me_range = hdi_middle_east['hdi'].max() - hdi_middle_east['hdi'].min()
me_cv = hdi_middle_east['hdi'].std() / hdi_middle_east['hdi'].mean()

print("South Asia HDI Range:", sa_range, "CV:", sa_cv)
print("Middle East HDI Range:", me_range, "CV:", me_cv)
# Insight: Higher CV indicates more variation in HDI within that region.
```

```python
# Load the regional datasets first
south_asia_df = pd.read_csv("HDI_SouthAsia_2020_2022.csv")
middle_east_df = pd.read_csv("HDI_MiddleEast_2020_2022.csv")
```

```python
# Task 6: Correlation Analysis


# Pearson correlation
sa_corr_gender = south_asia_df['hdi'].corr(south_asia_df['gender_development'])
sa_corr_life = south_asia_df['hdi'].corr(south_asia_df['life_expectancy'])

me_corr_gender = middle_east_df['hdi'].corr(middle_east_df['gender_development'])
me_corr_life = middle_east_df['hdi'].corr(middle_east_df['life_expectancy'])

print("South Asia - HDI vs Gender Development:", sa_corr_gender)
print("South Asia - HDI vs Life Expectancy:", sa_corr_life)
print("Middle East - HDI vs Gender Development:", me_corr_gender)
print("Middle East - HDI vs Life Expectancy:", me_corr_life)

# Scatter plots with trendlines
sns.lmplot(x='gender_development', y='hdi', data=south_asia_df)
plt.title("South Asia: HDI vs Gender Development")
plt.show()

sns.lmplot(x='life_expectancy', y='hdi', data=south_asia_df)
plt.title("South Asia: HDI vs Life Expectancy")
plt.show()

sns.lmplot(x='gender_development', y='hdi', data=middle_east_df)
plt.title("Middle East: HDI vs Gender Development")
plt.show()

sns.lmplot(x='life_expectancy', y='hdi', data=middle_east_df)
plt.title("Middle East: HDI vs Life Expectancy")
plt.show()
# Insight: Trendline direction and correlation values indicate strength and direction of relationships.

```

```python
# Task 7: Outlier Detection


def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    return outliers

# South Asia outliers
sa_hdi_outliers = detect_outliers(south_asia_df, 'hdi')
sa_gni_outliers = detect_outliers(south_asia_df, 'gross_inc_percap')

# Middle East outliers
me_hdi_outliers = detect_outliers(middle_east_df, 'hdi')
me_gni_outliers = detect_outliers(middle_east_df, 'gross_inc_percap')

print("South Asia HDI outliers:\n", sa_hdi_outliers[['country','hdi']])
print("South Asia GNI outliers:\n", sa_gni_outliers[['country','gross_inc_percap']])
print("Middle East HDI outliers:\n", me_hdi_outliers[['country','hdi']])
print("Middle East GNI outliers:\n", me_gni_outliers[['country','gross_inc_percap']])

# Scatter plots highlighting outliers
plt.figure(figsize=(10,6))
sns.scatterplot(data=south_asia_df, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=sa_gni_outliers, x='gross_inc_percap', y='hdi', color='red', label='Outliers')
plt.title("South Asia: HDI vs GNI per Capita (Outliers in Red)")
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(data=middle_east_df, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=me_gni_outliers, x='gross_inc_percap', y='hdi', color='red', label='Outliers')
plt.title("Middle East: HDI vs GNI per Capita (Outliers in Red)")
plt.show()

# Insight: Outliers may indicate countries whose GNI per capita is unusually high or low relative to their HDI.

```
